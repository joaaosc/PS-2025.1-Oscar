# Matriz de Correlação

Podemos também utilizar uma matriz de correlação, para tentarmos descobrir alguma relação entre as variáveis. Antes disso, retiramos as variáveis que são strings (identificadas anteriormente), para que a matriz de correlação possa ser calculada. 

```python
# Retirando as colunas string:

df = df.drop("Time 1", axis=1)
df = df.drop("Time 2", axis=1)
df = df.drop("Position 1", axis=1)
df = df.drop("Position 2", axis=1)
```

> OBS: Não salvaremos o data frame sem as strings aqui, pois elas ainda serão necessárias no futuro (para identificar dados duplicados).

```python
correlation_matrix = df.corr()

# Visualizar a matriz de correlação
plt.figure(figsize=(10, 10))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlação entre Variáveis do Conjunto de Dados MNIST')
plt.show()
```

Os buracos que aparecem no heatmap acima são consequência direta dos NaNs presentes no data frame.
